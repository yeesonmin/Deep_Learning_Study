```python
##### 경사 하강법 #####
"""
오차를 줄이는 방법
순간 기울기가 0인 지점을 찾는 것.
오차가 작은 기울기 a, y절편 b를 구하는것.
평균 제곱 오차( (예측 값 - 실제 값)^2의 합 / n)에서 a,b를 각각 편미분 한다.
"""

# 공부시간에 따른 예상 점수
#
#데이터 추가
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = [[2,81],[4,93],[6,91],[8,97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# 그래프로 나타내기
plt.figure(figsize = (8,8))
plt.scatter(x,y)
plt.show()

#리스트로 된 x,y를 배열로 바꾸기
x_data = np.array(x)
y_data = np.array(y)

#a, b 초기값
a = 0
b = 0

#학습률
learn = 0.005

#반복횟수
epochs = 2001

#경사하강법
for i in range(epochs):
    
    #y의 예측값
    y_pred = a*x_data + b
    error = y_data - y_pred

    #평균 제곱 오차를 a로 미분
    a_diff = (2 / len(x_data)) * sum(x_data*error)
    #평균 제곱 오차를 b로 미분
    b_diff = (2 / len(x_data)) * sum(error)

   
    #미분 결과와 학습률을 곱하여 a값 업데이트
    a = a + learn * a_diff
    b = b + learn * b_diff
    
    #출력
    if i % 100 ==0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f" % (i,a,b))

#그래프 다시 그리기
y_pred = a*x_data + b
plt.scatter(x,y)
plt.plot([min(x_data),max(x_data)],[min(y_data),max(y_data)])
plt.show
```
