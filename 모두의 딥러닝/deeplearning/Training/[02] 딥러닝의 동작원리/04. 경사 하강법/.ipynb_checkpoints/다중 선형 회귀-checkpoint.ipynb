```python
##### 다중 선형 회귀 #####
"""
여러개의 변수를 늘려서 정확하게 예측 할 수 있게 함.
"""

# 공부시간과 과외 횟수에 따른 예상 점수
#
#데이터 추가
data = [2,0,81],[4,4,93],[6,2,91],[8,3,97]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]

# 3D 그래프 그리는 라이브러리
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
ax = plt.axes(projection = '3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.scatter(x1,x2,y)
plt.show()

#리스트로 된 x,y를 배열로 바꾸기
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)

#a, b 초기값
a1 = 0
a2 = 0
b = 0

#학습률
lr = 0.005

#반복횟수
epochs = 8001

#다중 선형 회귀
for i in range(epochs) :
    #y의 예측값
    y_pred = a1 * x1_data + a2 * x2_data + b
    error = y_data - y_pred
    
    #평균 제곱 오차를 a로 미분
    a1_diff = -(2/len(x1_data))*sum(x1_data*error)
    a2_diff = -(2/len(x2_data))*sum(x2_data*error)
     #평균 제곱 오차를 b로 미분
    b_new = -(2/len(x1_data))*sum(error)
    
    #a,b값 업데이트
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_new
    
    #출력
    if i % 100 ==0:
        print("epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f" % (i,a1,a2,b))

ax = plt.axes(projection = '3d')
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.scatter(x1,x2,y)
plt.show()
```
