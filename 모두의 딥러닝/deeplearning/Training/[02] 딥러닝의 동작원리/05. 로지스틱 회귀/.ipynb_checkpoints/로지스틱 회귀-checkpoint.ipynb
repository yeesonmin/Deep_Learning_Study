```python
##### 로지스틱 회귀 #####
"""
참, 거짓 판단
시그모이드 함수 사용

기울기가 클수록 수직으로 커진다.
y절편에 따라 그래프가 좌우로 움직인다.

기울기와 y절편을 알기위해 경사하강법을 사용한다.
또한 로그 함수 사용해야 0,1을 구할 수 있다.
"""

#데이터 추가
import numpy as np
import pandas as pd
data =[2,0],[4,0],[6,0],[8,1],[10,1],[12,1],[14,1]

x_data=[i[0] for i in data]
y_data=[i[1]for i in data]

import matplotlib.pyplot as plt

plt.scatter(x_data,y_data)
plt.xlim(0,15)
plt.ylim(-1,2)
plt.show()

#a,b 초기화 및 학습률
a = 0
b = 0
lr = 0.05

#시그모이드
def sigmoid(x):
    return 1/(1+np.e**-x)

#a,b값 구하기
for i in range(2001):
    for x_data, y_data in data:
        #a편미분
        a_diff = x_data*(sigmoid(a*x_data + b) - y_data)
        #b편미분
        b_diff = sigmoid(a*x_data + b) - y_data
        #a,b업데이트
        a = a - lr*a_diff
        b = b - lr*b_diff
        
        if i % 100 == 0:
            print("echo=%.f,기울기=%.04f,절편=%.04f"%(i,a,b))
    
    
plt.scatter(x_data,y_data)
plt.xlim(0,15)
plt.ylim(-.1,1.1)
x_range = (np.arange(0,15,0.1)) #그래프로 나타낼 x값 범위
plt.plot(np.arange(0,15,0.1),np.array([sigmoid(a*x+b)for x in x_range]))
plt.show()
```
